```python
import matplotlib.pyplot as plt
from kmodes.kmodes import KModes

cost = []
K = range(1, 10)

for num_clusters in K:
    kmodes = KModes(n_clusters=num_clusters, init='Huang', n_init=5, verbose=1)
    kmodes.fit_predict(df)
    cost.append(kmodes.cost_)

plt.plot(K, cost, 'bx-')
plt.xlabel('Number of clusters')
plt.ylabel('Cost')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
from kmodes.kmodes import KModes
# Select specific columns for clustering
kmodes = KModes(n_clusters=3, init='Huang', n_init=5, verbose=1)
clusters = kmodes.fit_predict(df)

df['Cluster'] = clusters
print(df.head())
```

```python
print("Cluster Modes:")
print(kmodes.cluster_centroids_)
```

```python
import matplotlib.pyplot as plt

sns.countplot(x='Cluster', data=df)
plt.title('Cluster Distribution')
plt.show()
```

```python
import prince
```

```python
# Perform MCA
mca = prince.MCA(
    n_components=3,
    n_iter=3,
    copy=True,
    check_input=True,
    random_state=42
)

mca = mca.fit(df)
mca_result = mca.transform(df)

print(mca_result)
```

```python
plt.scatter(mca_result[0], mca_result[1], alpha=0.5)
plt.xlabel('Component 0')
plt.ylabel('Component 1')
plt.title('Scatter Plot of First two Components')
plt.show()
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3)
clusters = kmeans.fit_predict(mca_result)

# Add cluster labels to the DataFrame
mca_result['Cluster'] = clusters

# Plotting the clusters
plt.scatter(mca_result[0], mca_result[1], c=clusters, cmap='viridis', alpha=0.5)
plt.xlabel('Component 1')
plt.ylabel('Component 2')
plt.title('K-means Clustering on mca Components')
plt.show()
```

```python
centroids = kmeans.cluster_centers_
print(centroids)
```

```python
plt.scatter(mca_result[0], mca_result[1], mca_result[2], alpha=0.5)
plt.xlabel('Component 0')
plt.ylabel('Component 1')
plt.title('Scatter Plot of First three Components')
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(mca_result[0], mca_result[1], mca_result[2], alpha=0.5)
ax.set_xlabel('Component 1')
ax.set_ylabel('Component 2')
ax.set_zlabel('Component 3')
ax.set_title('3D Scatter Plot of First Three Components')
plt.show()
```
